```
df_A = gpd.read_file(root  + 'Cluster_Visualisation/' + str(7499.0) + '/MSOA_stab_df.shp')
df_B = gpd.read_file(root  + 'Cluster_Visualisation/' + str(3350.0) + '/MSOA_stab_df.shp')
df_C= gpd.read_file(root  + 'Cluster_Visualisation/' + str(188.4) + '/MSOA_stab_df.shp')
df_D = gpd.read_file(root  + 'Cluster_Visualisation/' + str(19.95) + '/MSOA_stab_df.shp')
df_E = gpd.read_file(root  + 'Cluster_Visualisation/' + str(0.5623) + '/MSOA_stab_df.shp')
```

```
ordered_df = pd.read_csv(root + 'ordered.csv')
ordered_df.head()
```

```
g = sns.jointplot(x=ordered_df[df_A['gi'] != 1]['HHI'], y=df_A[df_A['gi'] != 1]['gi'],kind='reg',joint_kws=dict(scatter_kws={'s':2},line_kws={'color':'red'}))
g.annotate(stats.pearsonr)
plt.suptitle('All MSOAs Correlation $g_{i}$ vs HHI')
plt.show()
```
